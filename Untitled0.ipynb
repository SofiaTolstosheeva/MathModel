**Простейшая модель Мальтуса роста населения Земли:**  y'= ky



```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def dydt(y, t):
    k=0.6
    return k*y
```

```python
t=np.linspace(0,10,100)
```

```python
solve=odeint(dydt,y0=1,t=t)
solve=solve.flatten()
%matplotlib inline
plt.plot(t, solve, label = r"$y'=-ky$",  color="yellow")
plt.grid(True)
plt.legend()
plt.show()
```
